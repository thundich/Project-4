```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path
from datetime import datetime

# File to Load
market_data_loading = Path("state_market_tracker.csv")

# Read csv and store into Pandas DataFrames
market_data= pd.read_csv(market_data_loading)
market_data
```

```python
#Dropping Columns in DataFrame that are not needed 
dropped_data= market_data.drop(columns=["period_duration", "region_type", "city", "median_sale_price_mom", "median_sale_price_yoy", "median_list_price_mom", "median_list_price_yoy", 
                  "median_list_ppsf_mom", "median_list_ppsf_yoy", "homes_sold_mom", "homes_sold_yoy", "pending_sales_mom", "pending_sales_yoy", "new_listings_mom", 
                  "new_listings_yoy", "inventory_mom", "inventory_yoy", "months_of_supply_mom", "months_of_supply_yoy", "median_dom_mom", "median_dom_yoy", 
                  "avg_sale_to_list_mom", "avg_sale_to_list_yoy", "avg_sale_to_list_mom", "avg_sale_to_list_yoy", "sold_above_list_mom", "sold_above_list_yoy", 
                  "price_drops", "price_drops_mom", "price_drops_yoy", "off_market_in_two_weeks", "off_market_in_two_weeks_mom", "off_market_in_two_weeks_yoy",
                  "last_updated", "parent_metro_region_metro_code", "is_seasonally_adjusted", "region_type_id", "region", "median_ppsf_mom", "median_ppsf_yoy", "median_dom"])
dropped_data
```

```python
#Splitting columns for period beginning
def convert(datetime_string_pb): 
    datetime_str= datetime.strptime(datetime_string_pb, "%m/%d/%Y")

    return datetime_str

pb_series = pd.Series(dropped_data["period_begin"])

year_array= []
month_array = []

for each in pb_series:
    pb_convert = convert(each)
    year_array.append(pb_convert.year)
    month_array.append(pb_convert.month)
    
dropped_data["year"]= year_array
dropped_data["month"] = month_array

market_data_updated= dropped_data.drop(columns=["period_begin", "period_end"])
market_data_updated
```

```python
#writing updated DataFrame to updated csv 
market_data_updated.to_csv("cleaned_state_market_tracker.csv")
```
