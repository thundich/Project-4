```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import distutils as _distutils
import tensorflow as tf
import tensorflow.compat.v1 as tf
from sklearn.metrics import balanced_accuracy_score, confusion_matrix, classification_report

# Import and read the cleaned_state_market_tracker.csv.
import pandas as pd 
cleaned_market_df = pd.read_csv("cleaned_state_market_tracker.csv")
cleaned_market_df.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
cleaned_market_df.drop(columns = ['table_id','rei_id', 'median_ppsf', 'property_type_id', 'state_code', 'median_list_ppsf', "avg_sale_to_list", "months_of_supply", "parent_metro_region", "year", "pending_sales"], inplace=True)
cleaned_market_df
```

```python
#erasing values that contain the word "null" in dataset 
null_application_df = cleaned_market_df.dropna(how='any', axis= 0)
null_application_df
```

```python
null_application_df['state'].value_counts()
```

```python
null_application_df['state'] = null_application_df['state'].astype('category')
null_application_df['state_code'] = null_application_df['state'].cat.codes
null_application_df= null_application_df.drop(columns= ['state'])
null_application_df
```

```python
# Separate the data into labels and features
# Separate the y variable, the labels
y = null_application_df["property_type"]

# Separate the X variable, the features
x = null_application_df.drop(columns="property_type")
```

```python
y.value_counts()
```

```python
# Split the data using train_test_split
# Assign a random_state of 1 to the function
x_train, x_test, y_train, y_test = train_test_split(x,y)
```

```python
# Instantiate the Logistic Regression model
# Assign a random_state parameter of 1 to the model
log_regress_model = LogisticRegression(random_state=1)

# Fit the model using training data
lr_model =log_regress_model.fit(x_train, y_train)
```

```python
# Make a prediction using the testing data
predict_testing_data = lr_model.predict(x_test)
predict_testing_data_df = pd.DataFrame({"Predictions": predict_testing_data, "Actual Result": y_test}).reset_index(drop=True)
predict_testing_data_df.head()
```

```python
# Print the balanced_accuracy score of the model
bal_accu_score = print(balanced_accuracy_score(y_test, predict_testing_data))
```

```python
# Generate a confusion matrix for the model
confusion_matrix(y_test, predict_testing_data)
```

```python
# Print the classification report for the model
class_report= classification_report(y_test, predict_testing_data)
print(class_report)
```
